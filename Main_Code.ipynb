```python
#Q learning aproach 

DState= [20] * 2 #making the q table with discrete size so that we can compute it since it will be hard almost infinity with the values mentioned from env
DSG = (env.observation_space.high - env.observation_space.low)/DState 
Q_intial = np.random.uniform(low=-2, high=0, size=(DState + [env.action_space.n]))#creating the q table of intial values below -2 as reward with the size of each state I will use the table is going to be 20*20*3

#Making the function to call when we need to return the tuple of disc state
def makedisc(state):
    d_state = (state - env.observation_space.low)/DSG
    return tuple(d_state.astype(np.int))
```

```python
EPISODES = 4000
Gamma = 0.95
LEARNING_RATE = 0.1
Show = 1000
epsilon = 1  
START_EPSILON_DECAYING = 1
END_EPSILON_DECAYING = EPISODES//2
epsilon_decay_value = epsilon/(END_EPSILON_DECAYING - START_EPSILON_DECAYING)



for episode in range(EPISODES):
    if episode_reward >-200:
            won+=1
    episode_reward = 0
    d_state = makedisc(env.reset())
    done = False   
    while not done:
        if np.random.random() > epsilon:
            # Use q table for an action after we already decayed the epsilon
            action = np.argmax(Q_intial[d_state])
        else:
            # Genrate random action until we learn for now
            action = np.random.randint(0, env.action_space.n)

        new_state, reward, done, info = env.step(action)
        episode_reward += reward
        new_d_state = makedisc(new_state)

      
        if not done:

            # Try to find the best Q value for the next step
            max_q = np.max(Q_intial[new_d_state])

            # Get the q of the current state and the action witth it 
            current_q = Q_intial[d_state + (action,)]

            # the q learning equation 
            new_q = (1 - LEARNING_RATE) * current_q + LEARNING_RATE * (reward + Gamma * max_q)

            # Updating the table with the new q value 
            Q_intial[d_state + (action,)] = new_q


        # If we reach the reward stage update the q value for that action 
        elif new_state[0] >= env.goal_position:
            Q_intial[d_state + (action,)] = 0

        d_state = new_d_state
    print(episode_reward)
    # Starting to decay the epsilon after a certain number of episodes 
    if END_EPSILON_DECAYING >= episode >= START_EPSILON_DECAYING:
        epsilon -= epsilon_decay_value
    
    
print ("won :" , won , "times")  
```
